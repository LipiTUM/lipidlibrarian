```python
import pandas as pd

from sqlalchemy import create_engine
from sqlalchemy.engine.reflection import Inspector
```

```python
engine = create_engine("mysql+pymysql://alex123:alex123@127.0.0.1:3306/alex123")
```

```python
inspector = Inspector.from_engine(engine)
table_names = inspector.get_table_names()
table_names
```

```python
for table_name in table_names:
	print(table_name)
	display(pd.read_sql_table(table_name=table_name, con=engine).head())
```
